```python
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=False), errors='coerce')
print(df['rating_count'].dtype)
```

```python
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
```

```python
df.info()
```

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
```

```python
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%', '').str.replace(',', ''), errors='coerce')
```

```python
df['rating'] = pd.to_numeric(df['rating'].astype(str), errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=False), errors='coerce')
```

```python
import matplotlib.pyplot as plt
```

```python
#Day 8Data Visualization
#bar chart
category_counts = df['category'].value_counts().head(10)

plt.figure(figsize=(12,8))
plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.title('Top 10 Product Categories by Count')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=90)
plt.show()
```

```python
#Line Chart
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.plot(category_rating.index, category_rating.values, marker='o', color='blue')
plt.title('Average Rating by Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
#Bar Chart
category_counts = df['category'].value_counts().head(5)

plt.figure(figsize=(7,7))
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.2f%%', startangle=90)
plt.title('Top 5 Category Share')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 15))  # 2 rows, 2 columns = 4 charts together

# Bar chart — top 10 categories by product count
category_counts = df['category'].value_counts().head(10)
axes[0, 0].bar(category_counts.index, category_counts.values, color='skyblue')
axes[0, 0].set_title('Top 10 Categories by Product Count')
axes[0, 0].tick_params(axis='x', rotation=90)

#  Line chart — average rating by category
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)
axes[0, 1].plot(category_rating.index, category_rating.values, marker='o', color='green')
axes[0, 1].set_title('Average Rating by Category')
axes[0, 1].tick_params(axis='x', rotation=90)

#  Pie chart — top 5 category share
top5 = df['category'].value_counts().head(5)
axes[1, 0].pie(top5.values, labels=top5.index, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Top 5 Category Share')

#  Bar chart — top 10 rated products
top_rated = df[['product_name','rating']].sort_values(by='rating', ascending=False).head(10)
short_names = top_rated['product_name'].str[:15] + '...'
axes[1, 1].bar(short_names, top_rated['rating'], color='orange')
axes[1, 1].set_title('Top 10 Rated Products')
axes[1, 1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

## Analysis
- The top categories (e.g., Electronics, Computers & Accessories) make up the largest share of products in the dataset.
- Average rating across top categories stays fairly consistent, mostly between 4.0 and 4.5.
- The top 5 categories account for a large portion of the total product catalog, shown in the pie chart.
- The highest-rated individual products stand out with ratings close to 5.0, though they may have fewer reviews than more popular categories.
