## 3. Data Cleaning

Several columns are stored as text (with `₹`, `,`, `%` symbols) instead of numbers. This section converts them to proper numeric types so they can be used in calculations.

```python
# Remove thousands-separator commas and convert rating_count to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=False), errors='coerce')
print(df['rating_count'].dtype)
```

```python
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
```

```python
df.info()
```

```python
# Remove ₹ symbol and commas so discounted_price can be used in calculations
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
```

```python
# Same cleanup as discounted_price
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
```

```python
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%', '').str.replace(',', ''), errors='coerce')
```

```python
df['discount_percentage'].head(10)
```

```python
df['rating'] = pd.to_numeric(df['rating'].astype(str), errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=False), errors='coerce')
```

```python
# Confirm there are no fully duplicated rows
df.duplicated().sum()
```

```python
# Double-check rating_count is now numeric
print(df['rating_count'].dtype)
```

```python
# Check for any rows with a missing product name
df[df['product_name'].isnull()]
```

## 6. Data Visualization

Charts to visualize category popularity, ratings, and market share.

```python
import matplotlib.pyplot as plt
```

```python
#Day 8 Data Visualization
#bar chart
category_counts = df['category'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(category_counts.index, category_counts.values, color='steelblue')
plt.title('Top 10 Product Categories by Count', fontsize=13)
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Line Chart
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.plot(category_rating.index, category_rating.values, marker='o', color='steelblue')
plt.title('Average Rating by Top 10 Categories', fontsize=13)
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Pie Chart
category_counts = df['category'].value_counts().head(5)

plt.figure(figsize=(7,7))
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Category Share', fontsize=13)
plt.tight_layout()
plt.show()
```

# Day 9 – Mini Dashboard

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

category_counts = df['category'].value_counts().head(10)
axes[0, 0].bar(category_counts.index, category_counts.values, color='steelblue')
axes[0, 0].set_title('Top 10 Categories')
axes[0, 0].tick_params(axis='x', rotation=45)

category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)
axes[0, 1].plot(category_rating.index, category_rating.values, marker='o', color='seagreen')
axes[0, 1].set_title('Average Rating by Category')
axes[0, 1].tick_params(axis='x', rotation=45)

top5 = df['category'].value_counts().head(5)
axes[1, 0].pie(top5.values, labels=top5.index, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Top 5 Category Share')

top_rated = df[['product_name','rating']].sort_values(by='rating', ascending=False).head(10)
axes[1, 1].bar(top_rated['product_name'].str[:15], top_rated['rating'], color='darkorange')
axes[1, 1].set_title('Top 10 Rated Products')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Day-11 Business Insights

1. **Category:** `Computers&Accessories|Accessories&Peripherals|Cables&Accessories|Cables|USBCables` is the highest buying category in the dataset, while `Electronics|Headphones,Earbuds&Accessories|Headphones|In-Ear` is the lowest buying category.

2. **Discount Percentage:** The maximum discount percentage is 94%, and the minimum is 0%, showing a wide variation in how products are discounted.

3. **Rating:** The maximum rating is 5.0 (best possible), and the minimum applied rating is 2.0.

4. **Average Rating:** The average rating across all products is 4.09, showing that most products are generally well-received by customers.

5. **Average Discount:** The average discount offered across all products is 46.7%, showing that discounting is a common practice across most listings, not just a few outlier products.

6. **Discount vs. Rating Relationship:** I checked whether higher discounts lead to lower ratings by calculating the correlation between `discount_percentage` and `rating`. The result was -0.16, which is very close to 0, meaning there is no real connection between the two — a product's discount percentage does not reliably predict its rating.

```python
print("Correlation between discount and rating:", round(df['discount_percentage'].corr(df['rating']), 2))
```
